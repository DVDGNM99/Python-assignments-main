# 2D molecule

```python


from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display

file_path = r"C:\Python-assignments-main\day04\chem-reporter\results\Caffeine\structure.sdf"


suppl = Chem.SDMolSupplier(file_path, removeHs=False)
mols = [m for m in suppl if m is not None]
assert mols, "Nessuna molecola valida trovata nel file SDF."

mol = mols[0]  
img = Draw.MolToImage(mol, size=(500, 400))
display(img)
```

# 3D molecule

```python


from rdkit import Chem
from rdkit.Chem import AllChem
import py3Dmol

file_path = r"C:\Python-assignments-main\day04\chem-reporter\results\Caffeine\structure.sdf"

suppl = Chem.SDMolSupplier(file_path, removeHs=False)
mols = [m for m in suppl if m is not None]
assert mols, "Nessuna molecola valida trovata nel file SDF."

mol3d = Chem.AddHs(mols[0])                 
AllChem.EmbedMolecule(mol3d, AllChem.ETKDG())  
AllChem.UFFOptimizeMolecule(mol3d)             

mb = Chem.MolToMolBlock(mol3d)

view = py3Dmol.view(width=520, height=420)
view.addModel(mb, "mol")
view.setStyle({"stick": {}})   # "cartoon", "line", "sphere"
view.zoomTo()
view.show()
```
